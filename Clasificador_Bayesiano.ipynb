Antes de pasar a la practica intentemos entender que es un clasificador bayesiano.

https://es.wikipedia.org/wiki/Clasificador_bayesiano_ingenuo

```python
class Bayes_class:
    import numpy as np
    
    
    def fit(self, X, y):
            
        self.means= {i: np.mean(X[y==i], axis=0) for i in y} #Diccionario con medias
        self.stds= {i: np.std(X[y==i], axis=0) for i in y}   #Diccionario con desviaciones standard          
        m=np.unique(y)
        self.props= {i : np.mean(y==m[i]) for i in y} 
        
        #print(self.means,self.stds,self.props)
        
    def prob(self,X, m, s):
        
        import scipy.stats as stats

        gaussian = [stats.norm.pdf(X[:,i],loc=m[i], scale=s[i]) for i in range (0,len(m))]
        #print(np.shape(gaussian))
        result=np.prod(gaussian, axis=0)
        return result
    
    def predict(self, X):
            
        clases = self.means.keys()
            
        pro_c = {clase:self.props[clase]*self.prob(X,self.means[clase],self.stds[clase]) for clase in clases}   
            
        maxi = np.zeros(len(X))
        predic = np.zeros(len(X))
            
        for clase in clases:
        
            arreglo = pro_c[clase]
            
            for j,pro in enumerate(arreglo):
                if pro>=maxi[j]:
                        
                    maxi[j]=pro
                    predic[j]=clase
            
       
        predictions = np.array(predic)
            
        return predictions
    
    def score(self, X, y):
        return np.mean(y==self.predict(X))
```
